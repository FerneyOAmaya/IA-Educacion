# Entrenar una red neuronal para clasificar texto

Emplearemos la herramienta Machine Learning For Kids a través del enlace https://machinelearningforkids.co.uk/.

Entrenaremos una red neuronal que pueda clasificar palabras entre dos posible grupos: Animales y Frutas.

##Paso 1: Crear los grupos

A continuación se presentan una tabla con palabras para cada uno de los dos grupos:

| Animales |  Frutas  |
|----------|----------|
| Ciervo   | Piña     |
| Cuervo   | Limón    |
| Hormiga  | Mango    |
| Gusano   | Uva      |
|Hipopótamo| Kiwi     |
| León     | Sandía   |
| Jirafa   | Melón    |
| Tigre    | Manzana  |
| Oso      | Banano   |
| Canguro  | Fresa    |

Emplearemos el 80% para entrenamiento y el 20% para pruebas

##Paso 2: Ingresar a Machine Learning For Kids
1. Ingresa a la página
2. Da click en el botón _Empezar_
3. Puedes registrarte o dar click en _Prueba sin registrarte_

##Paso 3: Crea el proyecto y las etiquetas
1. Selecciona _Añadir un nuevo proyecto_
2. Asigna lo siguiente, al finalizar dar click en _Crear_
- Nombre del proyecto: TextoIA
- Project type: Reconociendo Texto
- Idioma: Spanish
- Storage: In the cloud

##Paso 4: Entrena y prueba
1. Haz click en _Texto_, en _Entrenar_, crear las dos etiquetas y asigna las palabras de entrenamiento
2. En _Aprender y probar_, ingresar las palabras de prueba, una a la vez

## (Opcional) Paso 5: Emplea Colab y Maneja APIs

2. Revisa la instrucción para usa la API. El enlace debe finalizar de la siguiente forma:
 .../classify?data=Palabra

# (Opcional) Entrenar una red neuronal para clasificar imágenes

Emplearemos la herramienta Machine Learning For Kids a través del enlace https://machinelearningforkids.co.uk/.

Entrenaremos una red neuronal que pueda clasificar imágenes entre dos posible grupos: Caballos y Chacales.

##Paso 1: Preparar las imágenes

Previamente descarga un conjunto de imágenes (image dataset) del sitio Keggle (https://www.kaggle.com/). Nos interesa el conjunto de imágenes de animales mamíferos en el enlace: https://www.kaggle.com/datasets/asaniczka/mammals-image-classification-dataset-45-animals

Identifica las carpetas horse y kangooro.

Emplearemos 20 imagenes para entrenamiento.

##Paso 2: Ingresar a Machine Learning For Kids
1. Ingresa a la página
2. Da click en el botón _Empezar_
3. Puedes registrarte o dar click en _Prueba sin registrarte_

##Paso 3: Crea el proyecto y las etiquetas
1. Selecciona _Añadir un nuevo proyecto_
2. Asigna lo siguiente, al finalizar dar click en _Crear_
- Nombre del proyecto: ImagenesIA
- Project type: Reconociendo imágenes
- Storage: In the cloud

##Paso 4: Entrena y prueba
1.	Haz click en _Imágenes_, en _Entrenar_, crear dos etiquetas Caballo y Canguro (a la derecha Añadir etiqueta) con 20 imágenes por etiqueta. Debes arrastrar las imágenes desde Kaggle.
2.	Selecciona _Aprender & Probar_. En la parte inferior selecciona _Entrenar el modelo_. Luego añade una imagen para probar el modelo, usa la URL de la imagen.
