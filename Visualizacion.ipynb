# Looker Studio

Se presentará un ejercicio que realizaremos en Looker Studio de Google a manera de breve tutorial.

1. Registrarse e iniciar sesión en Looker Studio:
https://lookerstudio.google.com/

2. Crear un archivo en formato CSV con los siguientes datos:

<center>

|Producto	|Costo	|Tipo  |
|---------|-------|------|
|ebook|	100|	1|
|Libro|	10|	1|
|ebook|	200|	1|
|Revista|	1|	1|
|Libro|	20|	1|
|ebook|	300|	2|
|Revista|	2|	1|
|Libro|	30|	2|
|ebook|	400|	2|

</center>

3. Cargar la fuente de datos:
- Hacer clic en el icono *Crear* y luego seleccionar *Informe*.
- Seleccionar la pestaña *Conectarse a datos*. La pestaña *Mis fuentes de datos* incluye varias fuentes de datos de ejemplo.
- Seleccionar el conector *Subida de archivos*, pedirá autorización y debes seleccionar la fuente de datos, que es el archivo en formato CSV que se creó en el punto anterior.
- Selecciona el archivo y luego debes dar click en *Añadir* para añadir la fuente al informe.

4. Explorar el panel:
- Coloca un título (click en donde está *Informe sin título*) en la parte superior del panel.
- El lienzo está vacío
- En la parte derecha aparece una tabla con los campos de la fuente de datos.
- Se pudieran añadir más fuentes datos al informe en la parte inferior derecha.

5. Añadir campos, tablas y gráficos
Para añadir tablas y gráficos hay dos opciones:
- A través del menú con la opción Insertar
- Desde el botó del panel Añadir un gráfico.

6. Añadir campo: Añadiremos un campo que calcule la suma del costo y la llamaremos **Costo Sum**, este campo realizará agregación.
- En el menú de datos seleccionar *Añadir un campo*.
- Seleccionar *Añadir campo calculado*. Asignar el nombre **Costo Sum**.
- En fórmula colocar **SUM(Costo)**

7. Añadir cuadro de resultados: Añadiremos un cuadro de resultados para mostrar la suma total del Costo.
- Añadir el cuadro de resultados desde la opción del menú *Insertar* o desde el botón del panel *Añadir un gráfico*.
- En métrica seleccionar **Costo Sum**. Nos dará el valor resultante de sumar los valores de Costo.

8. Añadir Tabla: Añadiremos una tabla para visualizar todos los valores.
- Añadir una Tabla.
- En dimensión agregamos las columnas que queremos visualizar, en este caso, seleccionar las dimensiones: **Producto**, **Costo**, **Tipo**.
- Seleccionar la métrica **Costo Sum**.

9. Añadir otra Tabla: Añadiremos una tabla para agregar el costo por producto.
- Añadir una Tabla.
- En dimensión agregar las columnas que queremos visualizar, en este caso seleccionar la dimensión **Producto**.
- Seleccionar la métrica **Costo Sum**. Se observa como se agrega el costo por producto.

10. Añadir un control: Añadimos un control para controlar que tipo de productos se seleccionaran en todo el panel.
- Seleccionar *Añadir un control*, con la opción *Lista de tamaño fijo*.
- En *Campo de control* seleccionar *Tipo*.
- En este cuadro de control podemos seleccionar los tipos que queremos incluir. Cambiar esta selección para observar como cambian todas las tablas y el cuadro de resultados.

# Actividad

Realizaremos un panel de control más complejo a partir del conjunto de datos descargado de la Plataforma de Datos Abiertos de Colombia.

1. Crear un nuevo lienzo.

2. Añadir como fuente de datos la Tabla descargada de la Plataforma de Datos Abiertos de Colombia **MEN_INDICE_PARIDAD_POR_GENERO_MATRICULA_20240715**.

3. Añadir los siguientes campos:
- Total Masc: Suma todas las personas matriculadas hombres, la ecuación es: <BR>

<center>
SUM(MATR_MASC_PREJ)+SUM(MATR_MASC_TRANS)+SUM(MATR_MASC_PRIM)+SUM(MATR_MASC_SEC)+SUM(MATR_MASC_MEDIA)
<BR>
</center>

- Total Fem: Suma todas las personas matriculadas mujeres, la ecuación es:<BR>

<center>
SUM(MATR_FEM_PREJ)+SUM(MATR_FEM_TRANS)+SUM(MATR_FEM_PRIM)+SUM(MATR_FEM_SEC)+SUM(MATR_FEM_MEDIA)
<BR>
</center>

- Total Matricula: Suma todas las personas matriculadas hombres y mujeres, la ecuación es:<BR>

<center>
Total Fem+Total Masc
<BR>
</center>

4. Añadir control: Añadir el control tipo Lista de Tamaño Fijo para seleccionar los departamentos a procesar. Ordenar en forma Ascendente.

5. Añadir gráficos
Añadir los siguientes gráficos:

- Gráfico Circular:
  + Dimensión: Departamento
  + Métrica: Total Matrícula
- Gráfico de Barras: Gráfico de Columnas
  + Dimensión: Departamento
  + Métricas: Total Matrícula, Total Masc, Total Fem

6. Añadir cuadro de resultados
Añadir los siguientes cuadros de resultados:

- Total matrícula:
  + Métrica: Total Matrícula

- Total Masc:

    - Tipo de comparación: Métrica
    - Métrica objetivo: Total Matrícula
    - Mostrar como progreso: Seleccionado
- Total Fem:
  + Métrica: Total Fem
  + Otras opciones de comparación
    - Tipo de comparación: Métrica
    - Métrica objetivo: Total Matrícula
    - Mostrar como progreso: Seleccionado
